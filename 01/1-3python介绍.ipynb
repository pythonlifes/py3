# 一:认识python
    在python圈子中流行这么一句话,人生苦短，我用python(Life is short, you need Python)(Guido von Rossum,国内人称龟叔)
    

    编程语言的发明过程都是类似的，大部分都是起源于对于现状的不满，Python也不能免俗。吉多在设计Python之初，市面上已经存在有很多种不同的编程语言了，但似乎每种语言都不是那么的完美。C语言很强大，但是语法复杂，学习成本高，使用起来也比较麻烦。Shell很简单易用，但功能又不够强大，很多事都无法完成，吉多就想设计一款功能强大而又语法简洁的新语言。
    在1989年的圣诞节，吉多为了打发时间，开始着手编写新语言的解释器。这里我们提一下解释器，关于一门编程语言，可以简单的把它分成两个部分语法和解释器。语法指的是一些语法规则，各种语句啊，如何定义变量啊等等一系列的规则。而解释器的作用就是将符合这一系列规则的代码转换为机器码的程序。所以解释器程序开发完成，这门编程语言也就完成了。就这样，吉多利用一个圣诞节的时间开发了新语言的解释器。
    为什么叫python:
        当时电视上非常流行一个电视剧蒙提·派森的飞行马戏团。吉多也非常喜欢看，于是直接使用剧团的名字来为这个语言命名，Python由此诞生

# 二:什么是解释器(Cpython)
    1991年,第一个python解释器诞生,是用c语言实现的
    计算机不能直接理解除了机器语言以外的其他语言(eg:我们说的中文或者写的伪代码),因此需要把程序猿写的代码翻译成计算机认识可以执行的机器语言,程序才可以被编译执行。编译器就是将代码翻译成机器语言的工具( 源程序→编译器 →目标程序)
    编译型语言(c,c++):源代码-->编译器--->最终可执行文件 --->操作系统--->cpu
    解释型语言(python):源代码--->解释器(运行解释每一行代码)  --->操作系统--->cpu 
    
    运行速度和跨平台:
       解释型语言比编译型语言在跨平台方面好
       解释型语言在执行速度方面不如编译型语言
