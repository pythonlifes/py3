# 编码声明
    python程序由一个解析器读取,源码文件的文本编码可以自己指定声明
    编码解析流程:
        如果第一行注释位于python脚本的第一行或者第二行,并且通过正则coding[=:]\s*([-\w.]+)去匹配,编码声明必须单独一行,如果在第二行,第一行也必须是注释 # -*- coding: <encoding-name> -*-
# 字符串拼接
    1:显示拼接(以反斜杠结束的行不能带有注释):用\连接两行或者多行。
        if a < 1 and b > 2 \  # xxx 错误的写法
            c > 3:
            pass
        if a < 1 and b > 2 \
            c > 3:
            pass
    2:隐式拼接(可以带注释): 圆括号、方括号或花括号以内的表达式,无需使用反斜杠
        a = [1,2,3,4
            5,6]
# 缩进
    一个逻辑行开头处的空白 (空格符和制表符tab) 被用来计算该行的缩进等级，以决定语句段落的组织结构，一般的ide默认一个tab是4个空格
# 标识符和关键字
    1:标识符
        大写和小写字母 A 至 Z，下划线 _ 以及数字 0 至 9，但不可以数字打头，长度没有限制。对大小写敏感
        
    2:关键字
        import keyword
        keyword.kwlist
    3:保留的标识符(特殊含义)
        _:私有成员不会被from xx import导入,在交互式解释器中用来存放最近一次求值的结果
            a = 1
            b = 2
            _ = a
            _ 
            - = b
            _
        __*__: python标准库定义的,比如__init__(),__new()__,__call()__
        __* :表示类的私有名称
