# 字符串
    在 Python 中处理文本数据是使用 str 对象，也称为 字符串,由Unicode码位(https://baike.baidu.com/item/Unicode%E7%A0%81/7704811?fr=aladdin)构成的不可变 序列。 字符串字面值有多种不同的写法：

    单引号: '允许包含有 "双" 引号'
    双引号: "允许包含有 '单' 引号"。
    三重引号: '''三重单引号''', """三重双引号"""
    转义操作:
       print('C:\some\name')
       print('C:\some\\name')
       print(r'C:\some\name') # 前缀来禁用大多数转义序列的处理
     可以被索引(正向或者反向):
         a = '123'
         a[1] a[-1]
     api:
         str.capitalize() 
             返回原字符串的副本，其首个字符大写，其余为小写
         str.casefold()
            返回原字符串消除大小写的副本。 消除大小写的字符串可用于忽略大小写的匹配。
            与lower()类似,但是消除的更彻底 
                f = 'ß' 
                f.lower()
                f.casefold()
         str.center(width[, fillchar])
            返回长度为width的字符串，原字符串在其中间,width小于等于len(s) 则返回原字符串的副本
         str.count(sub[,start[,end]])
            返回子字符串sub在[start,end]范围内非重叠出现的次数 start起始为0
         str.encode(encoding="utf-8", errors="strict")
            指定的编码格式编码字节串对象的版本
         str.endswith(suffix[, start[, end]])
             判断字符串以什么结束,返回bool
             startswith()
             l.endswith('1',1,9)
         str.expandtabs(tabsize=8)
            返回字符串的副本，其中所有的制表符会由一个或多个空格替换
         str.find(sub[, start[, end]])
             rfind()
             返回子字符串sub在s[start:end] 切片内被找到的最小索引
         str.format(*args, **kwargs)
             字符串格式化
         str.index(sub[, start[, end]])
             rindex
            类似于 find()
         str.join(iterable)
             把可迭代对象拼接成字符串
         str.strip()
             rstrip()
             lstrip()
             删除空格
         str.replace(old, new[, count])
            返回字符串的副本，其中出现的所有子字符串 old 都将被替换为 new。 如果给出了可选参数 count，则只替换前 count 次出现
         str.split(sep=None, maxsplit=-1)
             rsplit()
             lsplit()
            返回一个由字符串内单词组成的列表，使用sep作为分隔字符串。如果给出了maxsplit则最多进行 maxsplit次拆分（列表最多会有 maxsplit+1 个元素），如果 maxsplit 未指定或为 -1，则不限制拆分次数（进行所有可能的拆分）。
         str.title()
            返回原字符串每个单词第一个字母为大写，其余字母为小写。
         str.upper()
             转大写
         str.zfill(width)
             填充字符串,如果 width 小于等于 len(s) 则返回原字符串的副本，正负值前缀 ('+'/'-') 的处理方式是在正负符号 之后 填充而非在之前

```python
a = 'pytHon'
b = a.capitalize()
print(b)
c = b.casefold()
print(c)

f = 'ß'
f.lower()
f.casefold()
```

```python
l = a.center(10,'1')
print(l)
```

```python
l.count('1',1,10) 
```

```python
l.encode()
```

```python
l.endswith('1',1,9)
```

```python
#如果\t前面的字符长度不足tabsize=4,缺几个就补几个空格,如果大于4,整除4后余数是几,按照4减去余数就是要补的空格数
'01\t012\t01234\t01\t012345'.expandtabs(tabsize=4) 
```

```python
l.find('z',2,12)
```

```python
'{}1{}'.format(1,1)
'{name}1{name}'.format(name='1')
'{name}1{age}'.format(name='2',age='1')
'{0}1{1}'.format('2','4')
```

```python
'{name} is python, value is {value} '.format(**{'name':'python','value':1})
```
